```python
# importing necessary packages

# ML_from_scratch imports
import sys
sys.path.append('../../src') # Temporarily adding the src folder to the path
from LinearRegression import LinearRegression

# standard python imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import normal
from sklearn.model_selection import train_test_split
```

## Generating dataset (1D)

A noisy line is created using numpy's normal function

```python
X = np.linspace(-5,5,300)
y = X + 10 + normal(scale=1.0, size=len(X))
```

Visualizing the data

```python
plt.scatter(X, y)
```

## Data preparation

I use sklearn.train_test_split() to split the data to testing and training sets.
Othen than that, there is not much data preparation needed as we have generated the data.
All the data is numerical and there are no missing data. The data is one dimensional so scaling is not required

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

## Fitting and predicting using ML_from_scratch's LinearRegression

### Using least squares method

```python
LinReg = LinearRegression(optimizer='least_squares')
```

```python
LinReg.fit(X_train,y_train)
```

```python
LinReg.RSS
```

```python
LinReg.R2
```

```python
LinReg.coeffs
```

### Using gradient descent

```python
LinReg = LinearRegression(optimizer='gradient_descent', learning_rate=0.0001, tolerance=10**-5)
```

```python
LinReg.fit(X_train,y_train)
```

```python
LinReg.RSS
```

```python
LinReg.R2
```

```python
LinReg.coeffs
```

Visualizing the prediction and compring to it to ground truth

```python
y_pred = LinReg.predict(X_test)
plt.plot(X_test, y_test, 'ro', alpha=0.5, label='Ground Truth')
plt.plot(X_test, y_pred, label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
LinReg.R2_score(X_test, y_test)
```

## Generating Dataset (2D)

```python
X = np.array([np.linspace(5,-5,300),np.linspace(0,10,300)]).transpose()
y = 2.0 + X[:,0] + 2*X[:,1]+ normal(scale=2.0, size=len(X[:,0]))
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].scatter(X[:,0], y)
ax[0].set_xlabel('X1')
ax[0].set_ylabel('y')

ax[1].scatter(X[:,1], y)
ax[1].set_xlabel('X2')
ax[1].set_ylabel('y')
```

## Data preparation

I use sklearn.train_test_split() to split the data to testing and training sets.
Othen than that, there is not much data preparation needed as we have generated the data.
All the data is numerical and there are no missing data. The data is one dimensional so scaling is not required

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

## Fitting and predicting using ML_from_scratch's LinearRegression

### Using least squares method

```python
LinReg = LinearRegression(optimizer='least_squares')
```

```python
LinReg.fit(X_train,y_train)
```

```python
LinReg.RSS
```

```python
LinReg.R2
```

```python
LinReg.coeffs
```

```python
y_pred = LinReg.predict(X_test)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 4))
ax[0].plot(X_test[:,0], y_test, 'ro', alpha=0.5, label='Ground Truth')
ax[0].plot(X_test[:,0], y_pred, label='Prediction')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('y')
ax[0].legend()

ax[1].plot(X_test[:,1], y_test, 'ro', alpha=0.5, label='Ground Truth')
ax[1].plot(X_test[:,1], y_pred, label='Prediction')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('y')
ax[1].legend()
```

```python
LinReg.R2_score(X_test, y_test)
```

### Using gradient descent

```python
LinReg = LinearRegression(optimizer='gradient_descent', learning_rate=0.00001, tolerance=10**-8)
```

```python
LinReg.fit(X_train,y_train)
```
